# **Contexto do Projeto**

O objetivo deste projeto é aplicar técnicas de aprendizado de máquina e análise estatística para explorar e modelar dois conjuntos de dados biológicos: a base de dados dos pinguins e a base de dados das flores de íris. Esses dados foram utilizados para construir modelos preditivos e classificatórios, além de realizar uma análise exploratória detalhada para entender melhor as relações entre as variáveis e suas implicações biológicas.

## 1\. Coleta de Dados

Primeiro, vamos importar as bibliotecas necessárias e carregar os dados de pinguins e flores de íris diretamente do Seaborn.

```python
# Importação das bibliotecas
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Carregar as bases de dados diretamente do Seaborn
# Base de Dados Pinguins
pinguins = sns.load_dataset("penguins")

# Base de Dados Iris
iris = sns.load_dataset("iris")

# Exibir as primeiras linhas dos datasets
print("Base de Dados Pinguins")
print(pinguins.head())
print("\nBase de Dados Iris")
print(iris.head())
```

## 2\.  Análise Exploratória de Dados (EDA)

Agora, vamos explorar as distribuições das variáveis e as relações entre elas.


### **2.1. Gráfico 1: Distribuição das variáveis dos pinguins**

O gráfico pairplot mostra como as variáveis numéricas se relacionam entre si e com a variável resposta (espécie).

```python
# Análise Exploratória Pinguins - Pairplot
plt.figure(figsize=(10, 6))
sns.pairplot(pinguins.dropna(), hue='species', diag_kind='kde')
plt.title('Distribuição das Variáveis dos Pinguins')
plt.show()
```

O pairplot permite visualizar como as variáveis contínuas se relacionam entre si e com a variável species. As distribuições ao longo da diagonal mostram a densidade das variáveis individuais. As cores indicam as diferentes espécies de pinguins.

### **2.2. **Gráfico 2: Boxplot do Comprimento do Bico por Espécie****





O gráfico boxplot ajuda a ver a dispersão e os possíveis outliers de uma variável, neste caso, o comprimento do bico.

```python
# Boxplot para observar o comprimento do bico por espécie
plt.figure(figsize=(10, 6))
sns.boxplot(x="species", y="bill_length_mm", data=pinguins)
plt.title('Boxplot - Comprimento do Bico por Espécie de Pinguim')
plt.show()
```

O boxplot mostra a distribuição do comprimento do bico (bill_length_mm) para cada espécie. Ele permite observar a mediana, os quartis, e possíveis outliers.

### **2.3. **Gráfico 3: Heatmap de Correlação entre as Variáveis dos Pinguins******



O heatmap exibe as correlações entre as variáveis numéricas dos pinguins, ajudando a identificar quais variáveis estão mais relacionadas.

```python
# Heatmap de correlação entre as variáveis numéricas dos pinguins
plt.figure(figsize=(10, 6))
corr_matrix = pinguins[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlação entre as Variáveis dos Pinguins')
plt.show()
```

O heatmap mostra como as variáveis estão correlacionadas. As variáveis com correlação mais forte aparecem com cores mais intensas. Por exemplo, o comprimento do bico e o comprimento da nadadeira podem mostrar correlação positiva.

## 3\. Pré-processamento dos Dado

Agora, vamos tratar os dados ausentes, padronizar as variáveis numéricas e transformar as variáveis categóricas para um formato numérico, o que é necessário para os modelos de Machine Learning.

```python
# Tratamento de Dados Faltantes
pinguins_clean = pinguins.dropna()  # Removendo as linhas com valores faltantes

# Padronização das variáveis numéricas
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
pinguins_clean[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']] = scaler.fit_transform(
    pinguins_clean[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]
)

# Transformação de variáveis categóricas em numéricas
pinguins_clean['species'] = pinguins_clean['species'].map({'Adelie': 0, 'Chinstrap': 1, 'Gentoo': 2})
pinguins_clean['sex'] = pinguins_clean['sex'].map({'Male': 0, 'Female': 1})

# Exibindo o resultado da transformação
print(pinguins_clean.head())
```

## 4\. Modelagem

### **4.1. Regressão Linear (Prever o Peso dos Pinguins)**

Vamos usar um modelo de Regressão Linear para prever o peso dos pinguins com base nas variáveis bill_length_mm, bill_depth_mm, e flipper_length_mm.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# Selecionar as variáveis preditoras (features) e a variável alvo (target)
X = pinguins_clean[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm']]
y = pinguins_clean['body_mass_g']

# Divisão entre treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Treinamento do modelo de Regressão Linear
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Previsões
y_pred = regressor.predict(X_test)

# Avaliação do modelo (RMSE)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE para o modelo de Regressão Linear: {rmse:.2f} g")
```

O RMSE (Root Mean Squared Error) é uma medida de erro que mostra o quão bem o modelo de regressão prevê os valores de peso dos pinguins. Um RMSE mais baixo indica melhor performance.

### **4.2 Árvore de Decisão (Classificar Espécies de Pinguins)**

Vamos usar uma Árvore de Decisão para classificar as espécies de pinguins com base nas variáveis físicas.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Seleção de variáveis preditoras (features) e a variável alvo (target)
X = pinguins_clean[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm']]
y = pinguins_clean['species']

# Divisão entre treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Treinamento do modelo de Árvore de Decisão
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

# Previsões
y_pred = clf.predict(X_test)

# Avaliação do modelo
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Acurácia para o modelo de Árvore de Decisão: {accuracy * 100:.2f}%")
print("Matriz de Confusão:")
print(conf_matrix)
```

### **4.3 K-Means (Agrupamento de Flores de Íris)**

Agora, vamos usar o algoritmo de K-Means para realizar o agrupamento das flores de íris com base em suas características físicas.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Pré-processamento das variáveis numéricas (padronização)
X_iris = iris.drop('species', axis=1)
scaler = StandardScaler()
X_iris_scaled = scaler.fit_transform(X_iris)

# Determinando o número ideal de clusters com o método do cotovelo
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(X_iris_scaled)
    wcss.append(kmeans.inertia_)

# Plot do método do cotovelo
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss)
plt.title('Método do Cotovelo')
plt.xlabel('Número de Clusters')
plt.ylabel('WCSS')
plt.show()

# Aplicar KMeans com o número ideal de clusters (3 clusters para íris)
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
y_kmeans = kmeans.fit_predict(X_iris_scaled)

# Visualização dos clusters
plt.figure(figsize=(10, 6))
plt.scatter(X_iris_scaled[y_kmeans == 0, 0], X_iris_scaled[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X_iris_scaled[y_kmeans == 1, 0], X_iris_scaled[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X_iris_scaled[y_kmeans == 2, 0], X_iris_scaled[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')
plt.title('Agrupamento das Flores de Íris com K-Means')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend()
plt.show()
```

O gráfico do método do cotovelo é usado para determinar o número ideal de clusters. O valor de WCSS (Within-Cluster Sum of Squares) mostra a compactação dos clusters. O ponto de inflexão geralmente indica o número ideal de clusters.

## 5\. Conclusões

*   **Regressão Linear:** O modelo apresentou um RMSE de aproximadamente 296g, o que representa cerca de 7% do peso médio dos pinguins, sugerindo uma boa previsão do peso.

*  **Árvore de Decisão:** O modelo alcançou uma acurácia de 96%, o que é excelente para classificar as espécies dos pinguins com base em suas características físicas.
